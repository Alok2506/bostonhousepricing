```python
X=data.drop(columns=['price'])
y=data['price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.2, random_state=0)
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
column_trans = make_column_transformer((OneHotEncoder(sparse_output=False),['beds']),remainder='passthrough')
```

```python
scaler = StandardScaler()
```

```python
lr = LinearRegression(normalize=True)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

#X being the feature matrix 
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

lr = LinearRegression()
lr.fit(X_scaled,y)
```

```python
pipe = make_pipeline(column_trans,scaler,lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred_lr = pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred_lr)
```

## Using Lasso

```python
lasso = Lasso()
```

```python
pipe = make_pipeline(column_trans,scaler,lasso)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred_lasso = pipe.predict(X_test)
r2_score(y_test,y_pred_lasso)
```

## Using Ridge

```python
ridge = Ridge()
```

```python
pipe = make_pipeline(column_trans,scaler,ridge)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred_ridge = pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred_ridge)
```

## using svm

```python
from sklearn.svm import SVR 
```

```python
svr_rbf = SVR(kernel = 'rbf', C=100, gamma=0.1,epsilon=0.1)
```

```python
pipe = make_pipeline(column_trans,scaler,svr_rbf)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred_svr = pipe.predict(X_test)
r2_score(y_test,y_pred_svr)
```

## using elastic net regression

```python
from sklearn.linear_model import ElasticNet
en = ElasticNet(alpha=1.0,l1_ratio=0.5, random_state=42)
pipe =make_pipeline(column_trans,scaler,en)
pipe.fit(X_train,y_train)
```

```python
y_pred_en = pipe.predict(X_test)
r2_score(y_test,y_pred_en)
```

```python
print("Linear Regression : ", r2_score(y_test,y_pred_lr))
print("Lasso : ", r2_score(y_test,y_pred_lasso))
print("Ridge : ", r2_score(y_test,y_pred_ridge))
print("SVM : ", r2_score(y_test,y_pred_svr))
print("ElasticNet :",r2_score(y_test,y_pred_en))
```
